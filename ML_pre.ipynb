```python
from jax import grad
import jax.numpy as jnp
import numpy as np

def func(x, y):
    return x**2*y + y**3*jnp.sin(x)

grad_jax = grad(func,argnums=[0,1])
grad_x,grad_y = grad_jax(0.0,1.0) # Using floating point values is necessary
print("Partial derivative wrt x at (0,1):",grad_x)
print("Partial derivative wrt y at (0,1):",grad_y)

def grad_check(x,y):
    grad_x = 2*x*y + y**3*np.cos(x)
    grad_y = x**2 + 3*y**2*np.sin(x)
    return grad_x, grad_y

print()
print("Checking values manually")
print("Partial derivative wrt x at (0,1):",grad_check(0,1)[0])
print("Partial derivative wrt y at (0,1):",grad_check(0,1)[1])
```

```python
import sympy as sp      #use SymPy for symbolic differentiation

x, y = sp.symbols('x y')

f = x**2 * y + y**3 * sp.sin(x)

gradient_x = sp.diff(f, x)
gradient_y = sp.diff(f, y)

# Print the gradients
print("Gradient with respect to x:", gradient_x)
print("Gradient with respect to y:", gradient_y)
```

```python
class Marks:
    def __init__(self, maths, english):
        self.maths = maths
        self.english = english

    def __repr__(self):
        return f"Maths: {self.maths}, English: {self.english}"


class Student:
    def __init__(self, roll_number, name, marks):
        self.roll_number = roll_number
        self.name = name
        self.marks = marks

    def __repr__(self):
        return (f"Roll Number: {self.roll_number}, Name: {self.name}, "
                f"Marks: {self.marks}")


class Branch:
    def __init__(self, branch_name):
        self.branch_name = branch_name
        self.students = []

    def add_student(self, student):
        self.students.append(student)

    def __repr__(self):
        students_info = "\n".join([f"  {student}" for student in self.students])
        return f"{self.branch_name}:\n{students_info}"


class Year:
    def __init__(self, year):
        self.year = year
        self.branches = []

    def add_branch(self, branch):
        self.branches.append(branch)

    def __repr__(self):
        branches_info = "\n".join([str(branch) for branch in self.branches])
        return f"Year: {self.year}\n{branches_info}"

database = []

# Adding records for 2022
year_2022 = Year(2022)
branch_1_2022 = Branch("Branch 1")
branch_1_2022.add_student(Student(1, "Alice", Marks(100, 70)))
branch_1_2022.add_student(Student(2, "Bob", Marks(85, 90)))

branch_2_2022 = Branch("Branch 2")
branch_2_2022.add_student(Student(3, "Charlie", Marks(95, 80)))

year_2022.add_branch(branch_1_2022)
year_2022.add_branch(branch_2_2022)

# Adding records for 2023
year_2023 = Year(2023)
branch_1_2023 = Branch("Branch 1")
branch_1_2023.add_student(Student(4, "David", Marks(88, 75)))

branch_2_2023 = Branch("Branch 2")
branch_2_2023.add_student(Student(5, "Eve", Marks(92, 85)))

year_2023.add_branch(branch_1_2023)
year_2023.add_branch(branch_2_2023)

# Adding records for 2024
year_2024 = Year(2024)
branch_1_2024 = Branch("Branch 1")
branch_1_2024.add_student(Student(6, "Frank", Marks(90, 78)))

branch_2_2024 = Branch("Branch 2")
branch_2_2024.add_student(Student(7, "Grace", Marks(85, 82)))

year_2024.add_branch(branch_1_2024)
year_2024.add_branch(branch_2_2024)

# Adding records for 2025
year_2025 = Year(2025)
branch_1_2025 = Branch("Branch 1")
branch_1_2025.add_student(Student(8, "Hannah", Marks(93, 80)))

branch_2_2025 = Branch("Branch 2")
branch_2_2025.add_student(Student(9, "Ivan", Marks(89, 77)))

year_2025.add_branch(branch_1_2025)
year_2025.add_branch(branch_2_2025)

# Add years to the database
database.extend([year_2022, year_2023, year_2024, year_2025])

# Print the database
for year in database:
    print(year)
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0.5, 100.5, 0.5)

functions = [
    (lambda x: x, 'y = x'),
    (lambda x: x**2, 'y = x^2'),
    (lambda x: x**3 / 100, 'y = x^3 / 100'),
    (lambda x: np.sin(x), 'y = sin(x)'),
    (lambda x: np.sin(x) / x, 'y = sin(x) / x'),
    (lambda x: np.log(x), 'y =log(x)'),
    (lambda x: np.exp(x), 'y = e^x')
]

for func, label in functions:
    y = func(x)
    
    plt.figure(figsize=(12, 6))
    plt.plot(x, y, label=label)
    plt.title(f'Plot of {label}')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.grid(True)
    plt.legend()
    
    # Special case for the exponential function to limit y-axis  (rapid growth of the exponential function, which can lead to extremely large values for large values of x)
    if label == 'y = e^x':
        plt.ylim([0, 10000])
    
    plt.show()
```

```python
import numpy as np

def argmin(array):
    if array.size == 0:
        return -1
        
    min_index = 0
    for i in range(1, array.size):
        if array[i] < array[min_index]:
            min_index = i
    
    return min_index

array = np.array([5, 2, 9, 1, 5, 6])

index = argmin(array)
numpy_index = np.argmin(array)

print(f"Manual argmin index: {index}")
print(f"Numpy argmin index: {numpy_index}")
```
